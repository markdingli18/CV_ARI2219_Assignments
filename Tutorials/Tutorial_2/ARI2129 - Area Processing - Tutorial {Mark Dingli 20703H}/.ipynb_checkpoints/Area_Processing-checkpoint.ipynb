##Convolution

An initial set of filters is preset for demonstration purposes.

```python
filters = [
    {'name': 'Identity', 'kernel': np.array([[0, 0, 0], 
                                             [0, 1, 0], 
                                             [0, 0, 0]])},
           
    {'name': 'Gaussian Blur', 'kernel': cv2.getGaussianKernel(ksize=5, sigma=2)},

    {'name': 'Sharpen', 'kernel': np.array([[0, -1, 0], 
                                            [-1, 5, -1], 
                                            [0, -1, 0]])},
           
    {'name': 'Edge Detection', 'kernel': np.array([[-1, -1, -1], 
                                                   [-1, 8, -1], 
                                                   [-1, -1, -1]])},
           
    {'name': 'Emboss', 'kernel': np.array([[-2, -1, 0], 
                                           [-1, 1, 1], 
                                           [0, 1, 2]])},
           
    {'name': 'Box Blur', 'kernel': np.array([[1, 1, 1], 
                                             [1, 1, 1], 
                                             [1, 1, 1]]) / 9},
]
```

The `filter2D` function is an OpenCV function that performs a 2D 
convolution of an image with a kernel or filter.

* The function takes three arguments: the input image, the depth of the output image, and the kernel to be used for convolution.
* The depth of the output image determines the bit-depth of the output image. This can be set to -1 to use the same depth as the input image, or to a specific value such as `cv2.CV_8U` to force an 8-bit unsigned output.
* The kernel argument should be a 2D matrix of floating-point or integer values. The size and values of the kernel determine the specific convolution operation being performed.
* This function returns a new image that is the result of the convolution operation.


```python
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(20, 10))
for i, f in enumerate(filters):
    filtered_img = cv2.filter2D(img, -1, f['kernel'])
    row = i // 3
    col = i % 3
    axes[row, col].imshow(filtered_img, cmap='gray')
    axes[row, col].set_title(f['name'])
    axes[row, col].axis('off')
```

```python
box_filters = [
    {'name': 'Identity', 'kernel': np.array([[0, 0, 0], 
                                             [0, 1, 0], 
                                             [0, 0, 0]])},
    {'name': 'Box Blur (3x3)', 'kernel': np.ones((3, 3), dtype=np.float32) / 9},
    {'name': 'Box Blur (5x5)', 'kernel': np.ones((5, 5), dtype=np.float32) / 25},
    {'name': 'Box Blur (7x7)', 'kernel': np.ones((7, 7), dtype=np.float32) / 49},
    {'name': 'Box Blur (15x15)', 'kernel': np.ones((15, 15), dtype=np.float32) / 225},
    {'name': 'Box Blur (25x25)', 'kernel': np.ones((25, 25), dtype=np.float32) / 625},
]
```

```python
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(20, 10))
for i, f in enumerate(box_filters):
    filtered_img = cv2.filter2D(img, -1, f['kernel'])
    row = i // 3
    col = i % 3
    axes[row, col].imshow(filtered_img, cmap='gray')
    axes[row, col].set_title(f['name'])
    axes[row, col].axis('off')
```

Differentiate between Gaussian and Box Blurring

```python
box_kernel = np.ones((5, 5), dtype=np.float32) / 25
gaussian_kernel = cv2.getGaussianKernel(ksize=5, sigma=0)

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))
filtered_img_box = cv2.filter2D(img, -1, box_kernel)
filtered_img_gaussian = cv2.filter2D(img, -1, gaussian_kernel)
axes[0].imshow(cv2.cvtColor(filtered_img_box, cv2.COLOR_BGR2RGB))
axes[0].set_title('Box Blur (3x3)')
axes[0].axis('off')
axes[1].imshow(cv2.cvtColor(filtered_img_gaussian, cv2.COLOR_BGR2RGB))
axes[1].set_title('Gaussian Blur (3x3)')
axes[1].axis('off')
```

Effect of different values for sigma in a Guassian filter

```python
img = cv2.imread(path,1)
#img= cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

#Define different values for sigma
sigmas = [1, 5, 10, 30]

# Apply each filter to the input image and display the result
fig, axes = plt.subplots(nrows=1, ncols=4, figsize=(20, 5))
for i, sigma in enumerate(sigmas):
    kernel = cv2.getGaussianKernel(ksize=3, sigma=sigma)
    filtered_img = cv2.filter2D(img, -1, kernel)
    axes[i].imshow(cv2.cvtColor(filtered_img, cv2.COLOR_BGR2RGB))
    axes[i].set_title('Gaussian Blur (3x3), sigma={}'.format(sigma))
    axes[i].axis('off')
```

Difference of Guassians

```python
img = cv2.imread(path,0)

# Define the sigma values for the Gaussian filters
sigmas = [1, 2]

# Apply each filter to the input image and display the result
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))
for i, sigma in enumerate(sigmas):
    kernel1 = cv2.getGaussianKernel(ksize=3, sigma=sigma)
    kernel2 = cv2.getGaussianKernel(ksize=3, sigma=sigma*2)
    kernel = kernel1 - kernel2
    filtered_img = cv2.filter2D(img, -1, kernel)

   #Image is inverted in this example to highlight the differences more.
    filtered_img_inverted = 255 - filtered_img
    axes[i].imshow(filtered_img_inverted, cmap='gray')
  
   # axes[i].imshow(filtered_img, cmap='gray')
    axes[i].set_title('DoG (3x3), sigma1={}, sigma2={}'.format(sigma, sigma*2))
    axes[i].axis('off')
```

Canny Edge Detector

```python
# Define the threshold values for the Canny edge detector
thresholds = [(50, 100), (100, 150), (150, 200), (200, 250)]

# Apply each threshold to the input image and display the result
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(15, 10))
for i, (threshold1, threshold2) in enumerate(thresholds):
    row = i // 2
    col = i % 2
    edges = cv2.Canny(img, threshold1, threshold2)
    axes[row][col].imshow(edges, cmap='gray')
    axes[row][col].set_title('Canny Edge Detection, threshold1={}, threshold2={}'.format(threshold1, threshold2))
    axes[row][col].axis('off')
```

Different Edge Detection Kernels

```python
ed_kernels = {
    'Identity': np.array([[0, 0, 0], [0, 1, 0], [0, 0, 0]]),
    'Sobel Top': np.array([[-1, -2, -1], [0, 0, 0], [1, 2, 1]]),
    'Sobel Bottom': np.array([[1, 2, 1], [0, 0, 0], [-1, -2, -1]]),
    'Sobel X': np.array([[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]]),
    'Sobel Y': np.array([[-1, -2, -1], [0, 0, 0], [1, 2, 1]]),
    'Sobel X+Y': np.array([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]]) + np.array([[-1, -1, -1], [0, 0, 0], [1, 1, 1]])
}
```

```python
# Apply each kernel to the input image and display the result
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(20, 10))
for i, (name, ed_kernels) in enumerate(ed_kernels.items()):
    row = i // 3
    col = i % 3
    filtered_img = cv2.filter2D(img, -1, ed_kernels)
    axes[row][col].imshow(filtered_img, cmap='gray')
    axes[row][col].set_title(name)
    axes[row][col].axis('off')
```
